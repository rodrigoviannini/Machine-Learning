```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pickle
```

```python
# importando e realizando a leitura das variaveis criadas no dataset de analise desses dados
with open('risco_credito.pkl', 'rb') as f:
  x_risco_credito, y_risco_credito = pickle.load(f)
```

```python
# Criando a arvore e a treinando (escolhendo o criterio que ela vai utilizar, Gini ou entropy)
arvore_risco_credito = DecisionTreeClassifier(criterion='entropy')
arvore_risco_credito.fit(x_risco_credito, y_risco_credito)
```

```python
# Vizualizando a ordem de importancia dada pela arvore aos atributos
arvore_risco_credito.feature_importances_
```

```python
# Vizualizando a arvore de decisão
tree.plot_tree(arvore_risco_credito); # Sem o ';' ele mostra em texto e na sequencia a imagem, para mostrar apenas a imagem é só adicionar o ';' no final.
```

```python
# Melhorando a vizualização
previsores = ['Historia', 'Divida', 'Garantia', 'Renda']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize = (10, 10))
tree.plot_tree(arvore_risco_credito, feature_names=previsores, class_names=arvore_risco_credito.classes_, filled=True);
```
